```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
print(dir(digits))
```

```python
#importing all models
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=500)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
```

```python
from sklearn.svm import SVC
svm = SVC()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
#implementing KFold cross validation there are 2 types
#stratified is better 
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10)
```

```python
from sklearn.model_selection import StratifiedKFold
skf = KFold(n_splits=10)
```

```python
#implementing cross validation with model
#this cross_val_score will automatically train and test and give the score
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(lr, digits.data, digits.target, cv = skf)
print(scores)

print(f"Mean accuracy: {scores.mean():.2f}")
```

```python
scores = cross_val_score(rf, digits.data, digits.target, cv = skf)
print(scores)

print(f"Mean accuracy: {scores.mean():.2f}")
```

```python
scores = cross_val_score(svm, digits.data, digits.target, cv = skf)
print(scores)

print(f"Mean accuracy: {scores.mean():.2f}")
```

```python
scores = cross_val_score(knn, digits.data, digits.target, cv = skf)
print(scores)

print(f"Mean accuracy: {scores.mean():.2f}")
```
