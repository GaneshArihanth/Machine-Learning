```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
# Load data
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
# Input and output
x = df[['year']]
y = df[['per capita income (US$)']]

# Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
# Create polynomial features (degree = 2, 3, or more if needed)
poly = PolynomialFeatures(degree = 2, include_bias = False)

x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
# Train model
reg = LinearRegression()
reg.fit(x_train_poly, y_train)
```

```python
print(reg.score(x_test_poly, y_test))
```

```python
# Plotting
%matplotlib inline
plt.scatter(df['year'], df['per capita income (US$)'], marker='.')
plt.show()
```

```python
# Generate smooth curve across the range of years
years = np.linspace(df['year'].min(), df['year'].max(), 100).reshape(-1, 1)
years_poly = poly.transform(years)
income_pred = reg.predict(years_poly)
```

```python
# Plot predictions
plt.scatter(df['year'], df['per capita income (US$)'], marker='.')
plt.plot(years, income_pred, color='red')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Per Capita Income Over Years with Polynomial Regression Curve')
plt.grid(True)
plt.show()
```
