```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
%matplotlib inline

plt.xlabel('Age')
plt.scatter(df[['age']], df[['bought_insurance']], marker='+', color='red')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df[['bought_insurance']], train_size=0.7, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
print(X_test)
reg.predict(X_test)
```

```python
reg.score(X_test,y_test) 
```

```python
import numpy as np
%matplotlib inline

plt.scatter(df.age,df.bought_insurance,marker='.',color='blue')

age = np.linspace(df['age'].min(), df['age'].max(), 10)

# To change 1D array to 2D array by using reshape(-1,1)
pred = reg.predict(age.reshape(-1, 1))

plt.plot(age, pred, color='red')

plt.xlabel('Age')
plt.show()
```
